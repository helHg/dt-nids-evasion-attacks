```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('ATTACKS.csv')

# Function to subtly modify features
def modify_features(row):
    if row['Label'] != 0:  # Only modify non-benign traffic
        # Adjust 'Flow Duration' to be closer to benign average
        row['Flow Duration'] = np.random.uniform(100000, 300000)
        
        # Reduce 'Tot Fwd Pkts' and 'Tot Bwd Pkts'
        row['Tot Fwd Pkts'] = max(2, row['Tot Fwd Pkts'] * 0.5)
        row['Tot Bwd Pkts'] = max(2, row['Tot Bwd Pkts'] * 0.5)
        
        # Adjust packet lengths
        row['TotLen Fwd Pkts'] = np.random.uniform(500, 1500)
        row['TotLen Bwd Pkts'] = np.random.uniform(500, 1500)
        
        # Modify flow IAT (Inter Arrival Time)
        row['Flow IAT Mean'] = np.random.uniform(5000, 15000)
        
        # Adjust packet rates
        row['Fwd Pkts/s'] = np.random.uniform(10, 100)
        row['Bwd Pkts/s'] = np.random.uniform(10, 100)
    
    return row

# Apply modifications
df_modified = df.apply(modify_features, axis=1)

# Save the modified dataset
df_modified.to_csv('ATTACKS_modified_features.csv', index=False)

print("Dataset with modified features has been saved.")
```
