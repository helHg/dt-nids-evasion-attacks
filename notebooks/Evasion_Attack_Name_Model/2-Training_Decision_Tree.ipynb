### 1- Making an 70/30 train/test split

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
import graphviz
import pydotplus
import sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree

# For reproducible results
```

```python
from sklearn.model_selection import train_test_split
RANDOM_STATE_SEED = 420
train, test = train_test_split(df_dataset, test_size=0.3, random_state=RANDOM_STATE_SEED)
```

#### MinMax Scaling of numerical attributes based on train set

```python
numerical_cols = df_dataset.columns[:-3]
numerical_cols
```

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler().fit(train[numerical_cols])
```

```python
train[numerical_cols] = min_max_scaler.transform(train[numerical_cols])
```

```python
test[numerical_cols] = min_max_scaler.transform(test[numerical_cols])
```

### 3- Splitting to X_train, y_train, X_test, y_test

```python
y_train = np.array(train.pop("Label")) # pop removes "Label" from the dataframe
X_train = train.values

print(type(X_train))
print(type(y_train))
print(X_train.shape)
print(y_train.shape)
```

```python
y_test = np.array(test.pop("Label")) # pop removes "Label" from the dataframe
X_test = test.values

print(type(X_test))
print(type(y_test))
print(X_test.shape)
print(y_test.shape)
```

### 4- Fitting Decision Tree model

```python
model = DecisionTreeClassifier(
    criterion='gini',
    splitter='best',
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0.0,
    max_features=None,
    random_state=None,
    max_leaf_nodes=None,
    min_impurity_decrease=0.0,
    class_weight=None,
    ccp_alpha=0.0
)

hyperparameters = {
    'max_depth': [i for i in range(1, 20)]
}
```

```python
clf = GridSearchCV(
    estimator=model,
    param_grid=hyperparameters,
    cv=5,
    verbose=1,
    n_jobs=-1  # Use all available CPU cores
)
```

```python
clf.fit(X=X_train, y=y_train)
```

### 5- Extracting best performing model in the 5-fold cross-validation Grid Search

```python
print("Accuracy score on Validation set: \n")
print(clf.best_score_ )
print("---------------")
print("Best performing hyperparameters on Validation set: ")
print(clf.best_params_)
print("---------------")
print(clf.best_estimator_)
```

```python
model = clf.best_estimator_
```

```python
model
```
